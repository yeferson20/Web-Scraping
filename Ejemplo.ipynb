## EJEMPLO 1 DE INVESTIGACION WEB SCRAPING:

Web Scraping de la pagina web "AS" que contiene datos deportivos, en este caso extraeremos los datos de la "LaLiga Santander" de los primeros 20 equipos en obtener la clasificacion y su respectivo puntaje

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Este código proporciona las herramientas necesarias para realizar solicitudes web, extraer información de una página web utilizando BeautifulSoup y trabajar con datos utilizando pandas. Podrías utilizarlo para obtener datos de una página web, analizarlos y realizar operaciones de análisis o manipulación de datos utilizando pandas.

```python
url = "https://resultados.as.com/resultados/futbol/primera/clasificacion/?omnil=src-sab"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

Obtenemos el contenido HTML de una página web, en este caso, la clasificación de la liga de fútbol española, utilizando requests. Luego, utiliza BeautifulSoup para analizar y estructurar ese contenido HTML, lo que permite extraer información específica de la página web más adelante.

```python
#EQUIPOS

eq = soup.find_all('span', class_="nombre-equipo")
equipos = []
count = 0
for i in eq:
    if count < 20:
        equipos.append(i.text)
    else:
        break
    count += 1
print(equipos, len(equipos))
```

```python
#Puntos

pt = soup.find_all('td', class_ ="destacado" )
puntos=list()
count = 0
for i in pt:
  if count < 20:
       puntos.append(i.text)
  else:
    break
  count += 1
print(puntos, len(puntos))
```

En esta parte del codigo, tanto para la seccion "EQUIPOS" y "Puntos" extraemos los nombres y la puntuacion de los primeros 20 equipos de la clasificación obtenida previamente a partir de la página web analizada con BeautifulSoup. Los nombres se almacenan en la lista equipos y luego se imprimen junto con la longitud de la lista.

Ahora teniendo los datos extraidos, los introduciremos en un Data Frame para poder usarlos de una forma mas facil.

```python
df = pd.DataFrame({'Nombre': equipos, 'Puntos':puntos}, index= list (range(1,21)))
print(df)
```

#### Teniendo ya el fragmento (que en este caso sería la tabla de los primeros 20 equipos clasificados) vamos a exportarlo en un archivo .cxv

```python
df.to_csv('Clasificacion.csv', index= False)
```

### EJEMPLO 3 DE INVESTIGACION WEB SCRAPING:

A continuacion tenemos una seleccion de peliculas de la pagina de IMDB que nos da una pelicula al azar de todo su catalogo para que podamos ver.

```python
import requests
from bs4 import BeautifulSoup
import random

# Realizar una petición GET a la página de IMDb con películas populares
url = 'https://www.imdb.com/chart/moviemeter/'
response = requests.get(url)

# Crear el objeto BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar todos los elementos de la tabla de películas
movies = soup.select('td.titleColumn')

# Seleccionar una película aleatoria
random_movie = random.choice(movies)

# Extraer el título y el año de la película seleccionada
title = random_movie.a.text
year = random_movie.span.text

# Imprimir el resultado
print('Película seleccionada:')
print('Título:', title)
print('Año:', year)
```

### EJEMPLO 4 DE INVESTIGACION WEB SCRAPING:

Pasos para implementar web scraping en python para extraer clasificaciones de películas IMDb y sus calificaciones:

Importe los módulos necesarios.

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

Acceda al contenido HTML de la página web asignando la URL y creando un objeto soap.

```python
# Downloading imdb top 250 movie's data
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

Extraiga las clasificaciones de las películas y sus detalles. Aquí, estamos extrayendo datos del objeto BeautifulSoup usando etiquetas HTML como href, title, etc.

```python
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
		for b in soup.select('td.posterColumn span[name=ir]')]
```

Después de extraer los detalles de la película, cree una lista vacía y almacene los detalles en un diccionario y, a continuación, agréguelos a una lista.

```python
# create a empty list for storing
# movie information
list = []

# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies)):

	# Separating movie into: 'place',
	# 'title', 'year'
	movie_string = movies[index].get_text()
	movie = (' '.join(movie_string.split()).replace('.', ''))
	movie_title = movie[len(str(index))+1:-7]
	year = re.search('\((.*?)\)', movie_string).group(1)
	place = movie[:len(str(index))-(len(movie))]
	data = {"place": place,
			"movie_title": movie_title,
			"rating": ratings[index],
			"year": year,
			"star_cast": crew[index],
			}
	list.append(data)
```

Ahora o la lista está llena de las mejores películas de IMBD junto con sus detalles. A continuación, muestre la lista de detalles de la película

```python
for movie in list:
	print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
		') -', 'Starring:', movie['star_cast'], movie['rating'])
```

Mediante el uso de las siguientes líneas de código, los mismos datos se pueden guardar en un archivo .csv se puede utilizar como un conjunto de datos.

```python
#saving the list as dataframe
#then converting into .csv file
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```

```python
prod=[]
prec=[]
```

```python
productos = s.findAll('div',{'class':'a-section octopus-pc-asin-title'})
precios = s.findAll('div',{'class':'a-section octopus-pc-asin-price'})

for  item in productos :
     prod.append(item.text.strip())
for  item in precios :
     prec.append(item.text.strip())

for i in range(len(prod)):
    print(prod[i], ':', prec[i])
```
